```python
import pandas as pd
```

```python
hospital = pd.read_csv("hospital.csv")
population = pd.read_csv("population.csv")
```

```python
hospital.tail()
# population.head()
```

```python
pop = population.drop(columns={'idxs','pop_18','pop_60','pop_12','pop_5'})
pop
```

```python
hospital = hospital.drop(columns={"beds","beds_noncrit",'admitted_pui','admitted_covid','admitted_total','discharged_pui','discharged_covid','discharged_total','hosp_covid','hosp_pui','hosp_noncovid'})
```

```python
merged = pd.merge(pop, hospital, on='state')
merged
```

```python
filterd_merged = merged.loc[merged['date'] == '2022-09-19']
filterd_merged
filterd_merged = filterd_merged.reset_index()
filterd_merged = filterd_merged.drop(columns={"index"})
filterd_merged
```

```python
filterd_merged.dtypes
```

```python
filterd_merged['state'] = filterd_merged['state'].replace(['W.P. Putrajaya','W.P. Kuala Lumpur','W.P. Labuan', 'Pulau Pinang'], ['Putrajaya','Kuala Lumpur', 'Labuan', 'Penang'])
```

```python
filterd_merged.to_csv("cleaned_hospital.csv")
```
